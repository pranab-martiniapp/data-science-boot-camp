## Questions

### 1 - Which teams made it to the post season?

```python
df = pd.io.gbq.read_gbq('''
  SELECT DISTINCT homeTeamName
  FROM `bigquery-public-data.baseball.games_post_wide`
  ''', project_id=project_id, dialect='standard'
)

df
```

### 2 - Regular Season Attendance Figures?

```python
df = pd.io.gbq.read_gbq('''
  SELECT DISTINCT
    g.homeTeamName,
    g.awayTeamName,
    s.attendance
  FROM
    `bigquery-public-data.baseball.games_wide` g
  JOIN
    `bigquery-public-data.baseball.schedules` s ON g.gameId = s.gameId
  ''', project_id=project_id, dialect='standard'
)

df.head()
```

### 3 - Team's total home game attendance?

```python
df = pd.io.gbq.read_gbq('''
  SELECT
    g.homeTeamName,
    SUM(s.attendance) AS TotalAttendance
  FROM
    `bigquery-public-data.baseball.games_wide` g
  JOIN
    `bigquery-public-data.baseball.schedules` s ON g.gameId = s.gameId
  GROUP BY
    g.homeTeamName
  ''', project_id=project_id, dialect='standard'
)

df
```
